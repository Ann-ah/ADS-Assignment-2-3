FACTORS THAT LEAD TO EMPLOYEE ATTRITION.

```python
sns.countplot(x = df["Age"])
plt.title("Frequency age distribution")
```

```python
sns.catplot(x="Attrition", y="Age", kind="bar", data=df)
plt.title("Bar plot of Age to Employee Attrition")
```

In regards to the plot above, age does not seem to be a major factor leading to employment attrition. However beyond the age of 35, according to the data, the attrition rate lessens. Also below the age of 30, employment attrition rate is high.

```python
sns.catplot(x="Attrition", y="RelationshipSatisfaction", kind="strip", data=df)
plt.title("Strip plot of Relationship satisfaction to Employee Attrition")
```

Regardless of the level of relationship satisfaction, employee attrition seems to be at the same rate.

```python
sns.catplot(x="Attrition", y="DailyRate", kind="box", data=df)
plt.title("Bar plot of DailyRate to Employee Attrition")
```

Daily minimum wage of 700 and below increased the rate employee attrition while a higher daily wage above 1000 lessened the rate of employee attrition.

```python
sns.catplot(x="Attrition", y="DistanceFromHome", kind="bar", data=df)
plt.title("Bar plot of Distance from home to Employee Attrition")
```

Distance from home does not seem to be a major factor contributing to employee attrition rate. However at a distance range of 9 and above the employee attrition rate is high. 

```python
sns.relplot(x = "Age", y = "TrainingTimesLastYear", hue ="Attrition", data = df, col = "BusinessTravel", row = "WorkLifeBalance")
```

Regardless of the age and worklife balance, employees that travelled rarely and the non-travellers retained their work contracts/ lessened the rate of employee attrition than those who travelled frequently. Nevertheless, those who had a higher worklife balance had a lower rate of attrition.

```python
sns.relplot(x = "YearsAtCompany", y = "YearsInCurrentRole", hue ="YearsWithCurrManager", data = df, col = "Attrition", row = "YearsWithCurrManager")
```

Regardless of the years in current role, the years with the current manager had a large significant effect in relation to employee attrition. The hypothesis is that the current manager offered a comfortable working environment which lessened employee attrition rate. Similarly, employees that had worked for the company for over 15 years had a lower rate of employee attrition. 

 Q1.Breakdown of distance from home by job role and attrition.

```python
df.groupby(['JobRole','Attrition']).DistanceFromHome.mean().unstack().plot(kind = "bar")
```

The higher the mean distance from home among various job roles the higher the employee attrition rates.

Q2.Compare average monthly income by education and attrition

```python
df["Monthly_Income"] = df["DailyRate"] * 30
Monthly_Income = df["Monthly_Income"]
Monthly_Income
```

```python
df.groupby(["Education","Attrition"]).Monthly_Income.mean().unstack().plot(kind = "bar")
```

Comparing the attrition rate in relation to education and mean monthly income: employees that had a lower level of education of below 2 had a slightly higher rate of attrition. The hypothesis is that the difference between the attrition rates is caused due to the employees leaving the workplace for greener pastures or education continuation. Those that had average education levels of 2-4, the attrition rate is lesser. the hypothesis is that the lower rate of attrition is caused by comfortability of the monthly income regardless of the education level. However education levels above five had a higher significant level of attrition rates. This may be due to unsatisfactory monthly mean wages in relation to the education level.
